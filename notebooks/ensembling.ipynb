```python
import pandas as pd
from pathlib import Path
```

```python
path = Path('../data/submissions')
our_submission = pd.read_csv(path / 'submission_405.csv') # 0.389
series2 = pd.read_csv(path / 'serie2.csv') # lgbm 0.379
series3 = pd.read_csv(path / 'serie3.csv') # autogluon 0.386
series4 = pd.read_csv(path / 'serie4.csv') # automl 0.383
series5 = pd.read_csv(path / 'serie5.csv') # xgb 0.369
sample_sub = pd.read_csv('../data/raw/sample_submission.csv')
```

```python
weights = {
    'our_submission': 0.7, # 'best_1dcnn.csv
    'serie2': 0.0,
    'serie3': 0.18,
    'serie4': 0.05,
    'serie5': 0.07
}

submission_blend = (
    our_submission['binds'] * weights['our_submission'] +
    series2['binds'] * weights['serie2'] +
    series3['binds'] * weights['serie3'] +
    series4['binds']  * weights['serie4'] +
    series5['binds']  * weights['serie5']
)

sample_sub['binds']=submission_blend
sample_sub.to_csv('submission.csv', index=False)
sample_sub
```
