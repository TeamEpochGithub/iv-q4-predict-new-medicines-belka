```python
import pandas as pd
from pathlib import Path
```

```python
path = Path('../data/submissions')
our_submission = pd.read_csv(path / 'best_1dcnn.csv')
series1 = pd.read_csv(path / 'serie2.csv')
series2 = pd.read_csv(path / 'serie3.csv')
series3 = pd.read_csv(path / 'serie4.csv')
series4 = pd.read_csv(path / 'serie5.csv')
sample_sub = pd.read_csv('../data/raw/sample_submission.csv')
```

```python
weights = {
    'serie2': 0.70,
    'serie3': 0.18,
    'serie4': 0.05,
    'serie5': 0.07
}

submission_blend = (
    our_submission['binds'] * weights['serie2'] +
    series2['binds'] * weights['serie3'] +
    series3['binds']  * weights['serie4'] +
    series4['binds']  * weights['serie5']
)

sample_sub['binds']=submission_blend
sample_sub.to_csv('submission.csv', index=False)
sample_sub
```
