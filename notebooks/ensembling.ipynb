```python
import pandas as pd
from pathlib import Path
```

```python
path = Path('../data/submissions')

# Our best submissions
our_submission = pd.read_csv(path / 'best_1dcnn.csv')

# External Submissions from https://www.kaggle.com/datasets/saberghaderi/semi-blending-2023/data
ext_lgbm = pd.read_csv(path / 'serie2.csv')  # LGBM Model: https://www.kaggle.com/code/mehrankazeminia/p-6-6-belka-competition-submission-lgbm
ext_autogluon = pd.read_csv(path / 'serie3.csv') # AutoGluon: https://www.kaggle.com/code/mehrankazeminia/5-belka-submission-autogluon-frag1-2-3
ext_automl = pd.read_csv(path / 'serie4.csv') # AutoML: https://www.kaggle.com/code/motono0223/leash-bio-automl-baseline
ext_xgboost = pd.read_csv(path / 'serie5.csv') # XGBoost: https://www.kaggle.com/code/ricopue/leashbio-xgb-ecfp-10m-sample-rows

# Sample Submission
sample_sub = pd.read_csv('../data/raw/sample_submission.csv')
```

```python
weights = {
    'our_submission': 0.3, # 'best_1dcnn.csv
    'serie2': 0.3,
    'serie3': 0.2,
    'serie4': 0.2,
    'serie5': 0.0
}

submission_blend = (
    our_submission['binds']     * 0.70 +
    ext_autogluon['binds']      * 0.18 +
    ext_automl['binds']         * 0.05 +
    ext_xgboost['binds']        * 0.07
)

sample_sub['binds']=submission_blend
sample_sub.to_csv('submission.csv', index=False)
sample_sub
```
