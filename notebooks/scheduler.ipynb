```python
from timm.scheduler.cosine_lr import CosineLRScheduler
from torch.optim import Adam
import torch
import matplotlib.pyplot as plt

# Create dummy parameters
dummy_params = [torch.nn.Parameter(torch.randn(2, 2, requires_grad=True))]

# Initialize optimizer with dummy parameters
optimizer = Adam(params=dummy_params, lr=1.0)

# Initialize CosineLRScheduler
scheduler = CosineLRScheduler(optimizer, t_initial=10, cycle_mul=3, cycle_decay=0.8, cycle_limit=5)

# List to store learning rates
lr = []

# Simulate epochs and collect learning rates
for epoch in range(100):
    scheduler.step(epoch)
    lr.append(optimizer.param_groups[0]['lr'])

# Plot learning rate over epochs
plt.plot(lr)
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Cosine Learning Rate Scheduler')
plt.show()
```
